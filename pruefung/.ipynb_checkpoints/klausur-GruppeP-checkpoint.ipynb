````python
#Konsistenzprüfung: Überprüfen, ob die Daten in verschiedenen Spalten konsistent sind

id_dept_mapping = {
"KU":"Kueche",
"IT":"IT",
"EW":"Entwicklung",
"MK":"Marketing",
"VT":"Verwaltung"
}

import re

# Ihre Lösung
regular_expression = "[A-Z]+" 
def check_id(id_, abteilung):
    dept_id = "".join(re.findall(regular_expression, id_))
    return id_dept_mapping.get(dept_id) == abteilung

# Ist der Check für alle Zeilen des Datensatzes erfolgreich?
df.apply(lambda x: check_id(x["id"], x["abteilung"]), axis=1).all()
````

> - *Eine **Umbenennung** von Spalten ist nicht notwendig!*

````python
# Ihre Lösung
jein_recoding = {
    'Y': 'Yes',
    'N': 'No'
}

df['interesse'] = df['interesse'].map(jein_recoding).fillna(df['interesse'])
````

````python
df['interesse'].value_counts()
````

````python
#einstellungs datum zur Datentzyp datetime tranformieren
df['einstellung'] = pd.to_datetime(df['einstellung'], format='%Y-%m-%d')
df['einstellung'].dtype
````

````python
#identische Duplikate löschen
duplicates = df.duplicated(subset=['id'], keep=False)
print("Anzahl der Duplikate in der Spalte 'ID':", duplicates.sum())
df[duplicates]
````

````python
#to remove all, keep=False
df.drop_duplicates(subset=['id'], keep='first', inplace=True)
df.duplicated(subset=['id'], keep=False).sum()
````

````python
#fehlende werte in spalte dienszeit mit Mean von Spalte einstellung ersetzen
import math

df['einstellung'] = pd.to_datetime(df['einstellung'])

df['dienstzeit'] = ((pd.to_datetime("2024-02-26", format="%Y-%m-%d") - df['einstellung']) / np.timedelta64(1,'Y')).apply(math.floor)

df['dienstzeit'] = df['dienstzeit'].apply(lambda x: 0 if x < 10 else (1 if x < 20 else 2))
````
