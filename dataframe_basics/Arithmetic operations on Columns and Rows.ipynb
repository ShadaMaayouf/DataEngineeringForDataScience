# Arithmetic operations on Columns and Rows

````python
import pandas as pd
df = pd.read_csv("./data/employees_satisfaction.csv", index_col=0)
df
````

### 1. Multiply a column by a coefficient
An example task would be to convert the values in the column ```salary```that are currently in US dollars to euros. 

> Note: Use the following exchange rate: 1$=0.9€

Usually the following operation is enough:

````python
df['salary'] *= 0.9
display(df['salary'])
````

However, all these options will work too:

``` df['salary'] = df['salary'].multiply(0,9)```

```df.loc[:, 'salary'] *= 0,9```

```df.iloc[:, index_of_column] = df.iloc[:, index_of_column]*0,9```
